```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pywt
from scipy import interpolate
import scipy.special as special
from scipy.integrate import quad
```

```python
def W_NGP(x, H):
    if np.abs(x)<(H*0.5):
        return 1.0/H
    elif np.abs(x)==(H*0.5):
        return 0.5/H
    else:
        return 0.0
    
W_NGP = np.vectorize(W_NGP)

def W_CIC(x, H):
    if np.abs(x)<H:
        return (1.0/H) - (np.abs(x)/(H*H))
    else:
        return 0.0
    
W_CIC = np.vectorize(W_CIC)

def W_TSC(x, H):
    if np.abs(x)<=(H*0.5):
        return (3.0/(4.0*H)) - ((x*x)/(H*H*H))
    elif (H*0.5)<np.abs(x)<=(1.5*H):
        return (0.5/H) * np.power( (1.5) - (np.abs(x)/H), 2 )
    else:
        return 0.0
    
W_TSC = np.vectorize(W_TSC)
```

```python
x = np.linspace(-2,2,num=500)

plt.plot(x, W_NGP(x, H=1), '--', label="NGP", linewidth=2)
plt.plot(x, W_CIC(x, H=1), '-.', label="CIC", linewidth=2)
plt.plot(x, W_TSC(x, H=1), ':',  label="TSC", linewidth=2)

plt.xlim(-2,2)
plt.ylim(-0.2,1.2)

plt.xticks([-2,-1.5,-1,-0.5,0,0.5,1,1.5,2])
plt.yticks([0,0.5,1])

plt.xlabel(r"$x\,[H]$", fontsize=20)
plt.ylabel(r"$W(x)$",   fontsize=20)

plt.legend(fontsize=14)

plt.savefig("./MAS_plots.pdf", bbox_inches='tight') 
plt.show()
```

```python
NGP, CIC, TSC = 1, 2, 3

def Wk(k,kN,p):
    return np.sinc( (0.5*k) / kN )**p

k = np.linspace(0,5,num=500)

plt.plot(k, (Wk(k, kN=1, p=NGP)), '--', label="NGP", linewidth=2)
plt.plot(k, (Wk(k, kN=1, p=CIC)), '-.', label="CIC", linewidth=2)
plt.plot(k, (Wk(k, kN=1, p=TSC)), ':',  label="TSC", linewidth=2)

plt.xlim(0,5)
plt.ylim(-0.4,1.2)

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,0.5,1])

plt.xlabel(r"$k\,[k_N]$", fontsize=20)
plt.ylabel(r"$W(k)$",   fontsize=20)

plt.legend(fontsize=14)

plt.savefig("./MAS_FT.pdf", bbox_inches='tight') 
plt.show()
```
